```python
for col in df.columns:
    if "Date" in col:
        df[col] = pd.to_datetime(df[col], errors="coerce")
print(df.dtypes)
```

```python
df.isnull().sum()
```

```python
df.drop(["OrderDate","PaymentID","PaymentDate","ReviewID","CustomerID_y","Rating","Comment","ReviewDate"],axis=1,inplace=True)
```

```python
 df["Amount"].fillna(0,inplace=True)
```

```python
 df["PaymentMethod"].fillna("Not Paid",inplace=True)
```

```python
df["Revenue"] = df["Price_x"] * df["Quantity"]
df.groupby("MenuItemID")["Revenue"].sum().nlargest(10)
```

```python
df.groupby("CuisineType")["OrderID"].nunique().reset_index().sort_values(by="OrderID", ascending=False)
plt.figure(figsize=(8,5))
plt.bar(df["CuisineType"], df["OrderID"])
plt.xticks(rotation=45)
plt.title("Top Cuisine Types by Number of Orders")
plt.xlabel("Cuisine Type")
plt.ylabel("Number of Orders")
plt.show()
```

```python
sns.histplot(df['Revenue'], kde = True, color ='blue', bins = 10)
```

```python
df["PaymentMethod"].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6,6))
plt.title("Payment Methods Distribution")
plt.show()
```

```python
df["PaymentMethod"].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6,6))
plt.title("Payment Methods Distribution")
plt.show()
```

```python
h= px.scatter( df, x="Quantity", y="Revenue", color="CuisineType", size="Revenue", hover_data=["FullName","Name"],title="Quantity vs Revenue by Cuisine Type")
h.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="CuisineType", y="Revenue")
plt.xticks(rotation=45)
plt.title("Revenue Distribution by Cuisine Type")
plt.show()
```

```python
orders = df.groupby(["CuisineType","PaymentMethod"])["OrderID"].nunique().reset_index()
fig = px.bar(orders, x="CuisineType", y="OrderID", color="PaymentMethod", title="Orders by Cuisine and Payment Method", barmode="stack",color_discrete_sequence=px.colors.sequential.Viridis)
fig.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[["Revenue","Amount","Quantity"]].corr(), annot=True, cmap="Blues")
plt.title("Correlation Heatmap")
plt.show()
```
